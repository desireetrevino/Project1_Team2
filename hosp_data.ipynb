```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import yfinance as yf
```

```python
# Define period
start = "2018-08-07"
end = "2023-08-07"
period = "1d"
```

## Hospitality Data Analysis

Name | Symbol
|----|-------|
HST | Host Hotels & Resorts Inc
HLT | Hilton Hotels Corporation
MAR | Marriott International Inc
CHH | Choice Hotels International Inc
H | Hyatt Hotels Corp
<!-- ABNB | Airbnb Inc -->

```python
# symbols = "HST HLT MAR ABNB CHH H"
symbols = "HST HLT MAR CHH H"
stocks_hist = pd.DataFrame({"Stock":[]})

stocks = yf.Tickers(symbols)
for ticker in stocks.tickers:
    df = stocks.tickers[ticker].history(start=start, end=end, period=period)
    df["Stock"] = ticker
    df["% change"] = round(100*(df["Close"] - df["Open"])/df["Open"], 2)
    df = df.reset_index()

    stocks_hist = pd.merge(stocks_hist, df, how="outer")
stocks_hist 
```

```python
# Export Stock Data to CSV
stocks_hist.to_csv("data/hosp.csv", index=False)
```
