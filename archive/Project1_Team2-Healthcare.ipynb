```python
# Dependencies:
import matplotlib.pyplot as plt
import pandas as pd
import requests
import datetime

# Import Yahoo Finance library to get Stock Market Data
import yfinance as yf
```

```python
# Get stocks from library
healthstocks = yf.Tickers("unh sny pfe mrna wba")
healthstocks_hist = healthstocks.history(start="2018-08-07", end="2023-08-07")
healthstocks_hist

```

```python
# Date vs Stock Value for each Stock:
UnitedHealth_Group_Incorporated = healthstocks_hist["Close"]["UNH"]
Sanofi = healthstocks_hist["Close"]["SNY"]
Pfizer_Inc = healthstocks_hist["Close"]["PFE"]
Moderna = healthstocks_hist["Close"]["MRNA"]
Walgreens_Boots_Alliance_Inc = healthstocks_hist["Close"]["WBA"]
UnitedHealth_Group_Incorporated.plot()
Sanofi.plot()
Pfizer_Inc.plot()
Moderna.plot()
Walgreens_Boots_Alliance_Inc.plot()
plt.legend(fontsize=7)
plt.axvline(datetime.datetime(2020, 1, 1), color='red', linewidth=1, linestyle='--')
plt.axvline(datetime.datetime(2021, 12, 1), color='red', linewidth=1, linestyle='--')
plt.show()
```

```python
# get the data
healthstocks_df = pd.DataFrame([{"Ticker":"","Date":"","Open":0,"High":0,"Low":0,"Close":0,"Adj Close":0,"Volume":0}])

stocks = ["unh","sny","pfe", "MRNA", "WBA"]
timeframe = "5y"
for stock in stocks:
    try:
        data = yf.download(stock, period=timeframe)
        data["Ticker"] = stock
        data = data.reset_index()
        healthstocks_df = healthstocks_df.append(data, ignore_index=True)
    except:
        print(f"{stock} not found")

fnastocks_df.head()
```
